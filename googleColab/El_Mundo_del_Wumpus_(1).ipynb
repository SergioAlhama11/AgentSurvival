```python
# TODO: Reemplaza YO! por tu nombre
agent_name = 'Estudiante'
import time


def move(self, message=''):
  """
  Esta función es la que hace que el agente decida el movimiento a realizar.

  ESTA ES LA FUNCIÓN QUE DEBES MODIFICAR A PARTIR DE LA PALABRA CLAVE TODO (que
  viene de "to do", es decir, "por hacer")

  No te preocupes por los argumentos que recibe esta función, ni por lo que
  debe devolver de salida
  """

  print(message, end='')
  messages = self.read_messages()


  whats_here = self.whats_here()

  objects = whats_here['objects']
  walls = whats_here['walls']

  for i in objects:
      if i['type'] == 'exit':
          i['exit_function'](self)
      if i['type'] == 'viento':
        self.turn_right()
        self.turn_right()
      if i['type'] == 'hedor':
        self.turn_right()
        self.turn_right()

  if walls['right'] == 0:
      self.turn_right()
      self.move_forward()
  elif walls['front'] == 0:
      self.move_forward()
  else:
      self.turn_left()

def move2(self, message=''):
    print(message, end='')
    messages = self.read_messages()

    whats_here = self.whats_here()
    objects = whats_here['objects']
    walls = whats_here['walls']
    for i in objects:
        if i['type'] == 'exit':
            i['exit_function'](self)
        if i['type'] == 'arrow':
            self.take_arrow()
        if i['type'] == 'stench':
            if walls['front'] == 0:
                self.shoot()

        if i['type'] == 'wind':
            self.turn_right()
            self.turn_right()
            if walls['right'] == 0 and walls['front'] == 0:
                self.move_forward()
                #print("OOO")
            if walls['right'] == 0:
                self.turn_left()
                #print("EEE")
                #self.stop()
                #self.move_forward()
    #self.stop()
    if walls['right'] == 0:
        self.turn_right()
        self.move_forward()
    elif walls['front'] == 0:
        self.move_forward()
    else:
        self.turn_left()

def move3(self, message=''):
    print(message, end='')
    messages = self.read_messages()

    whats_here = self.whats_here()

    objects = whats_here['objects']
    walls = whats_here['walls']

    if not walls['front']:
      self.move_forward()

    elif walls['front'] and not walls['right']:
      self.turn_right()

    else:
      self.turn_right()
```
